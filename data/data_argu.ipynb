```python
import os
import numpy as np
from PIL import Image

def crop(img, size, pos):
    ow, oh = img.size
    tw, th = size
    x1, y1 = pos        
    if (ow > tw or oh > th):        
        return img.crop((x1, y1, min(ow, x1 + tw), min(oh, y1 + th)))
    return img
img_path = "test.jpg"
img = Image.open(img_path)
display(img)
img2 = crop(img,(100,100),(20,20))
display(img2)
```
