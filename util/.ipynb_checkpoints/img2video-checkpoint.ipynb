```python
import cv2
import os
import numpy as np
from PIL import Image

def frame2video(im_dir,video_dir,fps): 

    im_list = os.listdir(im_dir)
    im_list = filter(lambda x: x.endswith(".jpg"), im_list)
    im_list = list(im_list)
    print(len(im_list))
    im_list.sort(key=lambda x: int(x.replace("output_SI_","").split('.')[0]))  #最好再看看图片顺序对不

    
    img = Image.open(os.path.join(im_dir,im_list[0]))
    img_size = img.size #获得图片分辨率，im_dir文件夹下的图片分辨率需要一致


    # fourcc = cv2.cv.CV_FOURCC('M','J','P','G') #opencv版本是2
    fourcc = cv2.VideoWriter_fourcc(*'XVID') #opencv版本是3
    

    videoWriter = cv2.VideoWriter(video_dir, fourcc, fps, img_size)

    # count = 1

    for i in im_list:
        #print(i)
        im_name = os.path.join(im_dir,i)
        
#         img = Image.open(os.path.join(im_dir,i))
#         img_size = img.size


        frame = cv2.imdecode(np.fromfile(im_name, dtype=np.uint8), -1)

        videoWriter.write(frame)


    videoWriter.release()

    

if __name__ == '__main__':
    
    
    fps = 24 #帧率，每秒钟帧数越多，所显示的动作就会越流畅
    save_dir = "/home/yangquanwei1/Motion_Transfer/C2F-FWN-new/results/solo_23_6/video" #合成视频存放的路径
    #save_dir = "/home/wujinlin5/yqw_home/Motion_Transfer/Evaluate/data/full_256p_5/video"
    os.makedirs(save_dir, exist_ok=True)
    video_dir = "/home/yangquanwei1/Motion_Transfer/C2F-FWN-new/results/solo_23_6/test_output_SI/"#帧存放路径
    #video_dir = "/home/wujinlin5/yqw_home/Motion_Transfer/Evaluate/data/full_256p_5/test_output_SI"
    video_list = os.listdir(video_dir)
    for video in video_list:
        if video[0:5]=='video': 
            frame_dir = os.path.join(video_dir,video)
            save_video_dir = os.path.join(save_dir,video+".mp4")
            print("frame_dir",frame_dir)
            print("save_video_dir",save_video_dir)

            frame2video(frame_dir, save_video_dir, fps)
    print('finish')

```

```python
import cv2
import os
import numpy as np
from PIL import Image

def crop(Ai):
    #print(Ai.shape)
    w = Ai.shape[1]
    base = 32
    x_cen = w // 2
    bs = int(w * 0.25) // base * base
    bs = 96
    return Ai[:,(x_cen-bs):(x_cen+bs),:]
    

imgs_path = "/home/yangquanwei1/Motion_Transfer/zeng/LWGAN/results_3/video19/imitators"
save_path = "/home/yangquanwei1/Motion_Transfer/zeng/LWGAN/results_3/video19/pred_new"
os.makedirs(save_path,exist_ok=True)


img_list = os.listdir(imgs_path)
img_list = filter(lambda x: x[0]=="p", img_list)
img_list = list(img_list)
    
    
img_list.sort(key=lambda x: int(x.replace("pred_","").split('.')[0])) 
for img_ in img_list[4:]:
    #print(img)
    img_path = os.path.join(imgs_path,img_)
    img = Image.open(img_path)
    #print(img.size)
    img_np = np.array(img)
    img_new_np = crop(img_np)
    img_new = Image.fromarray(img_new_np)
    new_path = os.path.join(save_path,img_)
    img_new.save(new_path)
    
```
